## Categorias de Usuário por Tipo de Atividade

-> **Categorias de usuário por tempo de tipos de atividades**

A Organização Mundial da Saúde (OMS) define atividade física como qualquer movimento corporal produzido pelos músculos esqueléticos que requer gasto de energia. A atividade física refere-se a todo movimento, incluindo durante o tempo de lazer, para transporte de ida e volta aos lugares, ou como parte do trabalho de uma pessoa. Tanto a atividade física de intensidade moderada quanto a vigorosa melhoram a saúde.

De Acordo com a OMS possui recomendações para diferentes faixas de idade. Neste caso, não existem dados sobre a idade, então será utilizada a faixa de adultos, que engloba pessoas entre 18 e 64 anos.

As recomendações para adultos são: 
* Deve realizar pelo menos **150 a 300 minutos** de atividade física aeróbica de intensidade **moderada**; ou pelo menos **75 a 150 minutos** de atividade física aeróbica de intensidade **vigorosa**; ou uma combinação equivalente de atividade de intensidade moderada e vigorosa ao longo da semana.

* Pode aumentar a atividade física aeróbica de intensidade **moderada** para mais de **300 minutos**; ou fazer mais de **150 minutos** de atividade física aeróbica de intensidade **vigorosa**; ou uma combinação equivalente de atividade de intensidade moderada e vigorosa ao longo da semana para benefícios adicionais à saúde. 

* Também deve realizar atividades de fortalecimento muscular com intensidade moderada ou maior que envolvam todos os principais grupos musculares em 2 ou mais dias por semana, pois estas proporcionam benefícios adicionais à saúde.

* Deve limitar a quantidade de tempo gasto em sedentarismo. Substituir o tempo sedentário por atividade física de qualquer intensidade (incluindo intensidade leve) proporciona benefícios à saúde, e

* Para ajudar a reduzir os efeitos prejudiciais de níveis elevados de comportamento sedentário na saúde, todos os adultos e idosos devem procurar fazer mais do que os níveis recomendados de atividade física de intensidade moderada a vigorosa.


fonte: https://www.who.int/news-room/fact-sheets/detail/physical-activity


Com essas informações é possivel rotular esses dados de maneira que:
* ( Atividade Moderada/Semana > 300 min OU Atividade Vigorosa/Semana > 150 min ) = Muito Ativo
* ( 150 min >= Atividade Moderada/Semana >= 300 min OU 75 min >= Atividade Vigorosa/Semana >= 150 min ) = Ativo
* Restante = Pouco Ativo ou Sedentário


<hr>

Observando os dados é possível ver que a primeira semana possui apenas 3 dias presentes, o que pode ocasionar em pequenas variações para realizar cálculos, já que não sabemos qual os valores do resto dos dias desta semana. Neste caso é bom procurar opiniões sobre como proseguir.

* Nesse caso decidi remover do calculo a primeira semana e focar nos 4 ultimas semanas de atividades registradas.
* Outra possibilidade é atribuir um peso diferente para a semana com menos registros disponíveis.

```r
weekly_activity <- daily_activity %>%
    filter(week != 15) %>%
    group_by(id, week) %>%
    summarise(total_active_minutes = sum(lightly_active_minutes + moderately_active_minutes + very_active_minutes),
        total_lightly_active_minutes = sum(lightly_active_minutes),
        total_moderately_active_minutes = sum(moderately_active_minutes),
        total_very_active_minutes = sum(very_active_minutes),
        total_sedentary_minutes = sum(sedentary_minutes))

head(weekly_activity)
```

```r
mean_weekly_activity <- weekly_activity %>%
    group_by(id) %>%
    summarise(total_active_minutes = mean(total_active_minutes),
        total_lightly_active_minutes = mean(total_lightly_active_minutes),
        total_moderately_active_minutes = mean(total_moderately_active_minutes),
        total_very_active_minutes = mean(total_very_active_minutes),
        total_sedentary_minutes = mean(total_sedentary_minutes))

head(mean_weekly_activity)
```

Com as regras definidas no início do tópico, pode-se supor que a atividade vigorosa equivale ao dobro de intensidade, ou seja, metade do tempo. Então é possivel calcular um meio termo com a equação $(x * 2) + y > 300$ (ou $x + (y / 2) > 150$) sendo x = atividade vigorosa e y = atividade moderada.

```r
user_type <- mean_weekly_activity %>%
    mutate(
        user_activity_level = case_when(
            ((total_moderately_active_minutes > 300) | (total_very_active_minutes > 150) | ((total_very_active_minutes * 2) + total_moderately_active_minutes) > 300) ~ "Very Active",
            ((total_moderately_active_minutes >= 150 & total_moderately_active_minutes <= 300) | (total_very_active_minutes >= 75 & total_very_active_minutes <= 150) | ((total_very_active_minutes * 2) + total_moderately_active_minutes) >= 150 & ((total_very_active_minutes * 2) + total_moderately_active_minutes) <= 300) ~ "Active",
            TRUE ~ "Slightly active or Sedentary"
        )
    ) %>%
    select(id, user_activity_level)

head(user_type)
```

```r
user_types <- user_type %>%
    group_by(user_activity_level) %>%
    summarise(
        count = n()
    )

user_types$user_activity_level <- factor(user_types$user_activity_level, levels = c("Slightly active or Sedentary", "Active", "Very Active"))

View(user_types)
```

```r
ggplot(user_types, aes(x = user_activity_level, y = count, fill = user_activity_level, label = count)) +
    geom_bar(stat = "identity", color = "black") +
    theme_minimal() +
    coord_flip() +
    scale_fill_manual(values = c("Slightly active or Sedentary" = "lightsalmon", "Active" = "lightblue", "Very Active" = "lightgreen")) +
    geom_label(position = position_stack(vjust = 0.5), size = 5) +
    labs(
        fill = "Atividade do Usuário",
        title = "Distribuição das Categorias de Usuário",
        x = "Categoria de Usuário",
        y = "Contagem",
        subtitle = "Quantidade de usuários por categoria de atividade."
    ) +
    theme( 
        axis.text.x = element_text(size = 15),
        axis.text.y = element_text(size = 15),
        axis.title = element_text(size = 15),
        plot.title = element_text(size = 20, face = "bold"),
        plot.subtitle = element_text(size = 15)
    ) +
    scale_y_continuous(
        breaks = seq(0, 14, 2)
    )
```

## Categorias de Usuário por Sono

De acordo com o artigo **"Recommended Amount of Sleep for a Healthy Adult: A Joint Consensus Statement of the American Academy of Sleep Medicine and Sleep Research Society"**:

* Adultos devem dormir 7 horas ou mais por noite regularmente para promover uma saúde ótima.

* Dormir menos de 7 horas por noite regularmente está associado a resultados adversos para a saúde, incluindo ganho de peso e obesidade, diabetes, hipertensão, doenças cardíacas e derrame, depressão e aumento do risco de morte. Dormir menos de 7 horas por noite também está associado a função imunológica prejudicada, aumento da dor, desempenho prejudicado, aumento de erros e maior risco de acidentes.

Fonte: https://jcsm.aasm.org/doi/10.5664/jcsm.4758

Com essas informações é possivel rotular esses dados de maneira que:

* Tempo de Sono >= 7 Horas = Sono Adequado
* Tempo de Sono < 7 Horas = Sono Inadequado

```r
daily_sleep <- daily_sleep %>%
    mutate(
        sleep_hours = round(total_minutes_asleep / 60, 2)
    )

head(daily_sleep)
```

Neste momento existe uma dificuldade que foi vista na limpeza dos dados, existem muitos valoes faltantes referentes ao sono dos usuários. O que pode prejudicar a análise, necessitando de mais dados.

```r
activity_records <- daily_activity %>%
    group_by(id) %>%
    summarise(
        activity_records_count = n()
    )

activity_records$use_type <- cut(activity_records$activity_records_count, 
                      breaks = c(-Inf, 1, 11, 21, Inf), 
                      labels = c("Not used", "Lightly used", "Regurlarly used", "Frequently used"),
                      right = FALSE)

head(activity_records)
```

* apenas 1% utiliza pouco durante o mês, variando entre 1 a 10 dias.
